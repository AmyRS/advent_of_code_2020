* How many bag colors can contain at least 1 shiny gold bag?
* Summarize Rules

```python
import re
```

Answer Below Part 1

```python
rule_input = []
data =  open(r"day_7_input.txt", "r").read().split('\n')
#data =  open(r"C:\Users\asorense\Downloads\day_7_test.txt", "r").read().split('\n')
for line in data:
    rule_input.append(line)
```

```python
bag_rule_dict = {}
for rule in rule_input:
    if len(rule) != 0:
        key = (re.match('(\A\w*\s\w*)', rule))[0]   
        matches = (re.findall('(\d\s\w*\s\w*)', rule))
        bag_rule_dict[key] = matches
```

```python
bag_rule_dict
```

```python
list_bags = []

def check_bags(list_of_bags, color):
    for item in bag_rule_dict:
        string = (bag_rule_dict.get(item))
        for bags in string:
            if color in bags:
                list_bags.append(item)
            
    for bags in list_of_bags:
        if 'shiny gold' in bags:
            list_of_bags.append(bags)
    return list_of_bags

check_bags(list_bags, 'shiny gold')

if list_bags:
    # loop for length of list to get each color and run through check
    len(list_bags)
    for bag_color in list_bags:
        #print(bag_color)
        check_bags(list_bags, bag_color)
    
else:
    print('null')
    
list_bags = set(list_bags)
len(list_bags)
```
